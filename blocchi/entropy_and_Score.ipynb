```python
def entropy(q_ij, s_ij):
    H=0
    lunghezza=len(q_ij.columns.value_counts())

    for j in range(0, lunghezza):
        H+=(q_ij.iloc[j,j:]*s_ij.iloc[j,j:]).sum()
    
    print H
    return H
```
