```python
from collections import Counter
import os
```

```python
alph_bear={'a':[], 'A':[], '=':[],
           'l':[], 'L':[], '^':[],
           'i':[], 'I':[], '+':[],
           'n':[], 'N':[], '>':[],
           's':[], 'S':[], '~':[],
           'b':[], 'B':[], '|':[],
           'y':[], 'Y':[], '@':[],
           '[':[], ':':[], '-':[]}
```

```python
lista=os.listdir("bear_new_alignment/")
```

```python
for famiglia in lista:
    f=open('bear_new_alignment/'+famiglia)

    line=f.readline()
    while(line):
        c=Counter(line.strip())
        #print c
        for chiave in c:
            alph_bear[chiave].append(c[chiave])
        #print alph_bear
        line=f.readline()

    #print alph_bear
```

```python
for chiave in alph_bear:
    alph_bear[chiave]=sum(alph_bear[chiave])
```

```python
import pandas as pd
```

```python
fr_attese=pd.DataFrame(columns=['a','A','=','l','L','^','i','I','+','n','N','>','s','S','~','b','B','|','y','Y','@','[', ':'],
                       index=['a','A','=','l','L','^','i','I','+','n','N','>','s','S','~','b','B','|','y','Y','@','[', ':'])

```

```python
fr_attese
```

```python
col=['a','A','=','l','L','^','i','I','+','n','N','>','s','S','~','b','B','|','y','Y','@','[', ':']

for index, row in fr_attese.iterrows():
    for colonna in col:
        if index==colonna:
            fr_attese.ix[index, colonna] = alph_bear[index]*2
        else:
            fr_attese.ix[index, colonna] = 2*alph_bear[index]*alph_bear[colonna]
        
```
