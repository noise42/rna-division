```python
f=open('train.cm').readlines()

pos=[]
name=[]
for i in range(0, len(f)):
    if f[i][0]=='I':
        pos.append(i)
        name.append(f[i+2].split()[1])
        
print name
print pos

for j in range(1, len(pos)):
    seq=f[pos[j-1]:pos[j]]
    i=j-1
    o=open('split_train/'+name[i]+'.cm', "w")
    
    for riga in seq:
        o.write(riga)
    o.close()

    
o=open('split_train/'+name[-1]+'.cm', "w")
seq=f[pos[-1]:]
for riga in seq:
    o.write(riga)
o.close()
```
