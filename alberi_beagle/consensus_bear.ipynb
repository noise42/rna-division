```python
f=open('Rfam_no_double.seed')
o=open('Rfam_no_double_consensus.fa', 'w')
line=f.readline()
while(line):
    if len(line.split())>=2:
        if line.split()[0]=='#=GF' and line.split()[1]=='AC':
            o.write('>'+line.split()[2]+"|")
        elif line.split()[0]=='#=GF' and line.split()[1]=='ID':
            o.write(line.split()[2]+"\n")
        elif line.split()[0]=='#=GC' and line.split()[1]=='RF':
            o.write(line.split()[2]+"\n")
        elif line.split()[0]=='#=GC' and line.split()[1]=='SS_cons':
            o.write(line.split()[2]+"\n")

    line=f.readline()
o.close()
```

```python
def decode_wuss(string_str, string_nt):
    new_str=""
    new_nt=''
    op=['(', '<', '[', '{']
    cl=[')', '>', ']', '}']
    for i in range(0, len(string_nt)):
        if string_nt[i] != ".":
            if string_str[i] in op:
                new_str+='('
            elif string_str[i] in cl:
                new_str+=')'
            else: 
                new_str+='.'
            new_nt+=string_nt[i].upper()
        
    return new_str, new_nt        
```

```python
def decode_nt(string):
    new=""
    for char in string:
        if char != ".":
            new+=char.upper()
    return new      
```
